```python
import pandas as pd
import re

def deep_clean_text(text):
    if pd.isna(text):
        return text

    text = str(text)

     # Odstrani HTML-podobne attribute npr. class="", style=""
    text = re.sub(r'\b(?:class|style|id|lang|dir)="[^"]*"', '', text)

    # Odstrani tage
    text = re.sub(r'\b/?(?:div|span|p|o:p)\b', '', text, flags=re.IGNORECASE)

     # Odstrani -- RICH TEXT --
    text = re.sub(r'--\s*RICH\s*TEXT\s*--', '', text, flags=re.IGNORECASE)

    # Odstrani <img ...>
    text = re.sub(r'<img[^>]*src="data:image[^"]*"[^>]*>', '', text, flags=re.IGNORECASE)

    # Odstrani "br" (case-insensitive)
    text = re.sub(r'\bbr\b', '', text, flags=re.IGNORECASE)

    # Odstrani vse < in >
    text = text.replace('<', '').replace('>', '')

    # Odstrani vse /
    text = text.replace('/', '')

    # Collapsea več presledkov
    text = re.sub(r'\s+', ' ', text).strip()

    return text

def clean_excel_file(input_path, output_path):
    excel = pd.ExcelFile(input_path)
    writer = pd.ExcelWriter(output_path, engine='openpyxl')

    for sheet in excel.sheet_names:
        df = excel.parse(sheet)
        df_cleaned = df.applymap(deep_clean_text)
        df_cleaned.to_excel(writer, sheet_name=sheet, index=False)

    writer.close()
    print(f"✅ Fully cleaned file saved to: {output_path}")


input_file = "FR_SR_WL_1.xlsx"
output_file = "cleaned_file.xlsx"
clean_excel_file(input_file, output_file)
```
